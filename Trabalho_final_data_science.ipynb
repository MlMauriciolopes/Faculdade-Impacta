```python
# Juntou as duas tabelas cadastro e salário
def salario(cadastro_merge):
    for cad in cadastro_merge['experiencia profissional']:
        if(cad <= 3):
            cadastro_merge['salario'] = cadastro_merge['salario até 3 anos']
        else:
            cadastro_merge['salario'] = cadastro_merge['salaior +3 anos']
    return cadastro_merge
```
